# Normalization

```python
#min-Max normalize the dataset
import numpy as np
def minmax_normalize(data):
    min_val = np.min(data, axis=0)
    max_val = np.max(data, axis=0)
   
    # Ensure the range is not zero to prevent division by zero
    range_val = np.where(max_val - min_val == 0, 1, max_val - min_val)
    
    normalized_data = (data - min_val) / range_val
    
    return normalized_data
```

```python
normailzed_numeric_data = minmax_normalize(encoded_merged_data[encoded_merged_data.columns.drop(["attack_cat"])])
normailzed_numeric_data
```

# Split The data and Test NB

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import CategoricalNB

#As data set contain both continuos data as well as Categorical Data, here calculating probability of each type 
# seperately and then combining it to find the accuracy. 


#Apply GuassianNB on continuous  features only


X = normailzed_numeric_data[normailzed_numeric_data.columns.drop(["proto","service","state","is_ftp_login","is_sm_ips_ports"])]
Y = df["attack_cat"]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=8)


guassianNB = GaussianNB()
clf_G = guassianNB.fit(X_train,y_train)
#y_predict = guassianNB.predict(X_test)
#print('GaussianNB Accuracy = ', accuracy_score(y_test,y_predict))
#Get probabality prediction 
train_probas_G = guassianNB.predict_proba(X_train)
test_probas_G = guassianNB.predict_proba(X_test)


#Apply CategoricalNB on categrical features only

X = normailzed_numeric_data[["proto","service","state","is_ftp_login","is_sm_ips_ports"]]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=8)

catagoricalNB = CategoricalNB()
catagoricalNB.fit(X_train,y_train)
#y_predict = catagoricalNB.predict(X_test)
#print('CategoricalNB Accuracy = ', accuracy_score(y_test,y_predict))
#Get probabality prediction 
train_probas_C = catagoricalNB.predict_proba(X_train)
test_probas_C = catagoricalNB.predict_proba(X_test)


#Combine probability prediction for both models

X_new_train = np.c_[train_probas_G,train_probas_C]
X_new_test = np.c_[test_probas_G,test_probas_C]

#Now apply GuassianNB on new X_new_train and X_new_test

new_gauassian = GaussianNB()
new_gauassian.fit(X_new_train,y_train)
y_predict = new_gauassian.predict(X_new_test)

print('Naive Bayes Accuracy = ', accuracy_score(y_test,y_predict))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression


#X = encoded_merged_data[encoded_merged_data.columns.drop(["attack_cat"])] #normailzed_numeric_data
X = normailzed_numeric_data
Y = df["attack_cat"]

model = LogisticRegression(penalty='l2',multi_class='multinomial')

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=8)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print('Logistic regression Accuracy = ',accuracy_score(y_test, y_pred))
```

# Feature Reduction Technique

```python
#Check with PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=15)
pca_full = pca.fit_transform(normailzed_numeric_data)

#plt.plot(np.cumsum(pca_full.explained_variance_ratio_))
#plt.xlabel('Number of components')
#plt.ylabel('Cumulative explained variance')


X_train, X_test, y_train, y_test = train_test_split(pca_full, Y, test_size=0.2, random_state=8)



guassianNB = GaussianNB()
guassianNB.fit(X_train,y_train)
y_predict = guassianNB.predict(X_test)

print('Naive Bayes Accuracy After PCA = ', accuracy_score(y_test,y_predict))
```

```python
#Forward Feature Selection with Naive Bayes
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.metrics import balanced_accuracy_score

max_score = 0
best_feature_selection_count = 0
for counter in range(1, X_train.columns.size):
    gaussianNB = GaussianNB()
    selector = SequentialFeatureSelector(gaussianNB, n_features_to_select=counter, direction='forward', scoring="balanced_accuracy", tol=0.01, cv=5)
    X_train_selected_feature = selector.fit_transform(X_train,y_train)
    gaussianNB.fit(X_train_selected_feature, y_train)

    X_test_selected_feature = selector.transform(X_test)

    y_pred = gaussianNB.predict(X_test_selected_feature)

    accuracy = balanced_accuracy_score(y_test, y_pred)
    print("Forward selection method Accuracy:", accuracy)
    if(max_score < accuracy):
        max_score = accuracy
        best_feature_selection_count = counter

print("Maximum accuracy :",max_score, " for ",best_feature_selection_count," feature selection")
```

```python
# LogisticRegression with Forward Feature Reduction 
max_score = 0
best_feature_selection_count = 0

X = normailzed_numeric_data
Y = df["attack_cat"]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=8)


for counter in range(1, X_train.columns.size):
    logisticRegression = LogisticRegression(penalty='l2',multi_class='multinomial')
    selector = SequentialFeatureSelector(logisticRegression, n_features_to_select=counter, direction='forward', scoring="balanced_accuracy", tol=0.01, cv=5)
    X_train_selected_feature = selector.fit_transform(X_train,y_train)
    logisticRegression.fit(X_train_selected_feature, y_train)

    X_test_selected_feature = selector.transform(X_test)

    y_pred = logisticRegression.predict(X_test_selected_feature)

    accuracy = balanced_accuracy_score(y_test, y_pred)
    print("Forward selection method Accuracy:", accuracy)
    if(max_score < accuracy):
        max_score = accuracy
        best_feature_selection_count = counter

print("Maximum accuracy :",max_score, " for ",best_feature_selection_count," feature selection")
```
