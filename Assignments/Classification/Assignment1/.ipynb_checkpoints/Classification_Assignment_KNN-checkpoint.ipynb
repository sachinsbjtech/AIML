# Outlier Analysis
There are outliers on each attribute so lets remove the outliers from the data set. Here outliers are in small number so can be dropped , othrwise different technique need to be used to get rid of outliers. 

```python
# Function to drop outliers from the atrributes
import numpy as np
def drop_outliers(df,field_name):
    iqr = 1.5 * ( np.percentile(df[field_name], 75) - np.percentile(df[field_name], 25))
    df.drop(df[df[field_name] > (iqr + np.percentile(df[field_name], 75))].index, inplace=True)
    df.drop(df[df[field_name] < (np.percentile(df[field_name], 25) - iqr)].index, inplace=True)
    #print("Field Name:",field_name,", Outlier count which is  > 1.5 * IQR:",df[df[field_name] > (iqr + np.percentile(df[field_name], 75))][field_name].count())
    #print("Field Name:",field_name,", Outlier count which is  < 1.5 * IQR:",df[df[field_name] < (np.percentile(df[field_name], 25) - iqr)][field_name].count())
```

```python
for f in df_processed.columns.drop(['Class']):
    drop_outliers(df_processed, f)
```

```python
# Divide the dataset into Attributes and Class 

X = df_processed[['Area', 'Perimeter', 'Major_Axis_Length', 'Minor_Axis_Length',
         'Eccentricity', 'Solidity', 'Extent',
       'Roundness','Aspect_Ration','Compactness']]
Y = df_processed['Class']
```

# Data Normalization 
Attribute data is numeric but some attribute values are large so need to nomalize the values. Using min-max normalization technique

```python
# normalize the dataset
import numpy as np

def minmax_normalize(data):
    min_val = np.min(data, axis=0)
    max_val = np.max(data, axis=0)
   
    # Ensure the range is not zero to prevent division by zero
    range_val = np.where(max_val - min_val == 0, 1, max_val - min_val)
    
    normalized_data = (data - min_val) / range_val
    
    return normalized_data


df_normalized = minmax_normalize(X)
```

```python
df_normalized
```

```python
#divide the data into Test and Train Data
from  sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df_normalized,Y,test_size=0.4,random_state=42)
```

```python
from scipy.stats import mode 

class KNN:
    def __init__(self,K):
        self.K = K
        
    def fit(self,x,y):
        self.X_train = x
        self.Y_train = y 
        self.x_train_number_of_examples, self.x_train_number_of_feature = x.shape
        
    def predict(self,x):
        self.x_test = x
        x_test_number_of_examples, x_test_number_of_feature = x.shape
        #received X_test data, lets predict Y for them
        #Initialize y_predict array with x_test size 
        
        #y_predict = np.zeros(x_test_number_of_examples)
        y_predict = [""] * x_test_number_of_examples
        
        #Classify each x_test example i.e. predict y value 
        for i in range(x_test_number_of_examples):
            x_test_example = self.x_test.iloc[i]
            
            
            #find euclidean distance of x_test example from all X_train examples
            
            #initialize euclidean_distance array of X_train examples size to zeros
            euclidean_distance = np.zeros(self.x_train_number_of_examples)
            euclidean_distance = np.sqrt(np.sum(np.square(x_test_example, self.X_train)))
            
            #Sort euclidean_distance
            indices = euclidean_distance.argsort()
            
            #get the respective Y_train values for those indices
            Y_train_sorted = self.Y_train.iloc[indices]
            
            #get first K items from Y_train_sorted which gives K nearest neighbour classes
            
            K_neighbours = Y_train_sorted[:self.K]
            
            #Find highest frequency class and store it
            class_value, counts = np.unique(K_neighbours,return_counts=True)
            #find highest frequency class and assign this as x_test example predicted value
            y_predict[i] =class_value[np.argmax(counts)]
            
        return y_predict
            
        
    def classification_score(self,y_true, y_pred):
        correct_predictions = 0
        for i in range(len(y_true)):
            if y_true.iloc[i] == y_pred[i]:
                correct_predictions +=1
        
        return (correct_predictions/len(y_true)*100)
```

```python
#Find the accuracy when k =3 
knn = KNN(3)
knn.fit(X_train,y_train)
y_predict = knn.predict(X_test)

print("KNN Model Accuracy when k=3 is :",knn.classification_score(y_test,y_predict))
```

```python
#Lets try to find the accuracy when k = [5,7,9]
k_accuracy_map = {}
for k in [5,7,9]:
    knn = KNN(k)
    knn.fit(X_train,y_train)
    y_predict = knn.predict(X_test)
    k_accuracy_map[k]=knn.classification_score(y_test,y_predict)

# print accuracy for respective K value
for key in k_accuracy_map:
    print("KNN Model Accuracy when k=",key," is :",k_accuracy_map[key])
```

```python
#Lets compare the accuracy for all ( even and odd ) K values. 
k_accuracy_map = {}
for k in range(1, 15):
    knn = KNN(k)
    knn.fit(X_train,y_train)
    y_predict = knn.predict(X_test)
    k_accuracy_map[k]=knn.classification_score(y_test,y_predict)

# print accuracy for respective K value
for key in k_accuracy_map:
    print("KNN Model Accuracy when k=",key," is :",k_accuracy_map[key])
```

```python
#Plot K-value and Accuracy graph
plt.title('k-NN Varying number of neighbors')
plt.plot(k_accuracy_map.keys(), k_accuracy_map.values(), label='Testing Accuracy')
#plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.xticks(range(1, 20))
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```
